```python
#model 설계
#========================================================================================================================================================#
class CRNN(nn.Module):
    def __init__(self, num_chars, img_height=32, img_width=100):
        super(CRNN, self).__init__()
        # num_chars: 실제 문자 개수(+2) -> 여기서는 +2(CTC Blank 등)를
        # LabelConverter에서 이미 blank를 붙여줬으니,
        # 파이토치의 CTCLoss는 blank 인덱스를 지정 가능하므로, +2 없이 설계할 수 있음.
        # 여기서는 blank를 0번으로 하므로 굳이 +2가 필요 없을 수 있습니다.
        # 필요시엔 +1 혹은 +2로 조정하세요.
        self.num_chars = num_chars

        # (3, H, W) -> (64, H, W)
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool2d(2, 2)  # (64, H/2, W/2)

        # (64, H/2, W/2) -> (128, H/2, W/2)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(2, 2)  # (128, H/4, W/4)

        # (128, H/4, W/4) -> (256, H/4, W/4)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.conv4 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.pool3 = nn.MaxPool2d((1, 2))  # (256, H/4, W/8)

        # (256, H/4, W/8) -> (512, H/4, W/8)
        self.conv5 = nn.Conv2d(256, 512, kernel_size=3, padding=1)
        self.bn5 = nn.BatchNorm2d(512)
        self.conv6 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.bn6 = nn.BatchNorm2d(512)
        self.pool4 = nn.MaxPool2d((1, 2))  # (512, H/4, W/16)

        # (512, H/4, W/16) -> (512, (H/4)-1, (W/16)-1) conv(2,2)
        self.conv7 = nn.Conv2d(512, 512, kernel_size=(2, 2))
        # 최종 (512, (H/4)-1, (W/16)-1)
        # 예) H=32 -> H/4=8 -> (8)-1=7
        #     W=100 -> W/16=6.25 -> 실제 integer shape는 계산 필요
        # 텐서플로우 예시에선 (24, 512) 시퀀스로 reshape했으니, 여기선 실제 연산으로 확인 필요

        # Bi-LSTM
        self.lstm1 = nn.LSTM(512, 256, bidirectional=True, batch_first=True)
        self.lstm2 = nn.LSTM(512, 256, bidirectional=True, batch_first=True)

        # 최종 fc
        self.fc = nn.Linear(512, self.num_chars)

    def forward(self, x):
        # (B,3,32,100)
        x = F.relu(self.conv1(x))      # -> (B,64,32,100)
        x = self.pool1(x)             # -> (B,64,16,50)
        x = F.relu(self.conv2(x))      # -> (B,128,16,50)
        x = self.pool2(x)             # -> (B,128,8,25)
        x = F.relu(self.conv3(x))      # -> (B,256,8,25)
        x = F.relu(self.conv4(x))      # -> (B,256,8,25)
        x = self.pool3(x)             # -> (B,256,8,12) (25->12)
        x = F.relu(self.conv5(x))      # -> (B,512,8,12)
        x = self.bn5(x)
        x = F.relu(self.conv6(x))      # -> (B,512,8,12)
        x = self.bn6(x)
        x = self.pool4(x)             # -> (B,512,8,6)
        x = F.relu(self.conv7(x))      # -> (B,512,7,5) (8->7, 6->5)

        b, c, h, w = x.size()
        # 시퀀스 길이 = h*w
        x = x.view(b, c, h * w)  # (B,512,35)
        x = x.permute(0, 2, 1)   # (B,35,512)

        # LSTM
        x, _ = self.lstm1(x)    # (B,35,512)
        x, _ = self.lstm2(x)    # (B,35,512)

        # 최종 FC
        x = self.fc(x)          # (B,35,num_chars)

        # PyTorch의 CTCLoss를 위해선 (T,B,C) 형태가 일반적
        # 여기서는 (B,T,C) -> (T,B,C)
        x = x.permute(1, 0, 2)  # (35,B,num_chars)
        return x

class CRNN_karas(nn.Module):
    def __init__(self, num_chars, img_height=32, img_width=100):
        super(CRNN_karas, self).__init__()
        # num_chars: 실제 문자 개수(+2) -> 여기서는 +2(CTC Blank 등)를
        # LabelConverter에서 이미 blank를 붙여줬으니,
        # 파이토치의 CTCLoss는 blank 인덱스를 지정 가능하므로, +2 없이 설계할 수 있음.
        # 여기서는 blank를 0번으로 하므로 굳이 +2가 필요 없을 수 있습니다.
        # 필요시엔 +1 혹은 +2로 조정하세요.
        self.num_chars = num_chars

        # (3, H, W) -> (64, H, W)
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool2d(2, 2)  # (64, H/2, W/2)

        # (64, H/2, W/2) -> (128, H/2, W/2)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(2, 2)  # (128, H/4, W/4)

        # (128, H/4, W/4) -> (256, H/4, W/4)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.conv4 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.pool3 = nn.MaxPool2d((1, 2))  # (256, H/4, W/8)

        # (256, H/4, W/8) -> (512, H/4, W/8)
        self.conv5 = nn.Conv2d(256, 512, kernel_size=3, padding=1)
        self.bn5 = nn.BatchNorm2d(512)
        self.conv6 = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.bn6 = nn.BatchNorm2d(512)
        self.pool4 = nn.MaxPool2d((1, 2))  # (512, H/4, W/16)

        # (512, H/4, W/16) -> (512, (H/4)-1, (W/16)-1) conv(2,2)
        self.conv7 = nn.Conv2d(512, 512, kernel_size=(2, 2))
        # 최종 (512, (H/4)-1, (W/16)-1)
        # 예) H=32 -> H/4=8 -> (8)-1=7
        #     W=100 -> W/16=6.25 -> 실제 integer shape는 계산 필요
        # 텐서플로우 예시에선 (24, 512) 시퀀스로 reshape했으니, 여기선 실제 연산으로 확인 필요

        self.fc_pre_lstm = nn.Linear(512, 64)

        # Bi-LSTM
        self.lstm1 = nn.LSTM(64, 256, bidirectional=True, batch_first=True)
        self.lstm2 = nn.LSTM(512, 256, bidirectional=True, batch_first=True)

        # 최종 fc
        self.fc = nn.Linear(512, self.num_chars)

    def forward(self, x):
        # (B,3,32,100)
        x = F.relu(self.conv1(x))      # -> (B,64,32,100)
        x = self.pool1(x)             # -> (B,64,16,50)
        x = F.relu(self.conv2(x))      # -> (B,128,16,50)
        x = self.pool2(x)             # -> (B,128,8,25)
        x = F.relu(self.conv3(x))      # -> (B,256,8,25)
        x = F.relu(self.conv4(x))      # -> (B,256,8,25)
        x = self.pool3(x)             # -> (B,256,8,12) (25->12)
        x = F.relu(self.conv5(x))      # -> (B,512,8,12)
        x = self.bn5(x)
        x = F.relu(self.conv6(x))      # -> (B,512,8,12)
        x = self.bn6(x)
        x = self.pool4(x)             # -> (B,512,8,6)
        x = F.relu(self.conv7(x))      # -> (B,512,7,5) (8->7, 6->5)

        b, c, h, w = x.size()
        # 시퀀스 길이 = h*w
        x = x.view(b, c, h * w)  # (B,512,35)
        x = x.permute(0, 2, 1)   # (B,35,512)
        
        x = self.fc_pre_lstm(x)  # (B,35,64)
        
        # LSTM
        x, _ = self.lstm1(x)    # (B,35,512)
        x, _ = self.lstm2(x)    # (B,35,512)

        # 최종 FC
        x = self.fc(x)          # (B,35,num_chars)

        # PyTorch의 CTCLoss를 위해선 (T,B,C) 형태가 일반적
        # 여기서는 (B,T,C) -> (T,B,C)
        x = x.permute(1, 0, 2)  # (35,B,num_chars)
        return x
```

```python
# 시각화 및 보조 함수/ 기타
#=================================================================================================================================================================================
def Loss_Visualization(train_losses, val_losses, model_name="Model", mode1="Training", mode2="Validation"):
    plt.plot(range(len(train_losses)), train_losses, 'b-',label=f'{model_name} {mode1} Loss')
    plt.plot(range(len(val_losses)), val_losses,'r--', label=f'{model_name} {mode2} Loss')
    plt.legend()
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title(f"{model_name} {mode1} and {mode2} Loss")
    plt.show()

def Edit_Distance_Visualization(train_edit_distance, val_edit_distance, model_name="Model", mode1="Training", mode2="Validation"):
    plt.plot(range(len(train_edit_distance)), train_edit_distance, 'b-',label=f'{model_name} {mode1} Edit Distance')
    plt.plot(range(len(train_edit_distance)), val_edit_distance,'r--', label=f'{model_name} {mode2} Edit Distance')
    plt.legend()
    plt.xlabel("Epochs")
    plt.ylabel("Edit Distance")
    plt.title(f"{model_name} {mode1} and {mode2} Edit Distance")
    plt.show()

def unnormalize(img_tensor):
    """
    img_tensor: [C, H, W] 텐서 (전처리 상태, 예: normalization 적용됨)
    ImageNet 평균 및 표준편차를 이용하여 복원 (RGB 순서)
    """
    if img_tensor.dim() == 4:
        img_tensor = img_tensor[0]
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])
    img = img_tensor.cpu().numpy().transpose(1, 2, 0)
    img = std * img + mean
    img = np.clip(img, 0, 1)
    img = np.uint8(255 * img)
    return img

class LabelConverter(object):

     def __init__(self, character):
         self.character = "-" + character
         self.label_map = dict()
         for i, char in enumerate(self.character):
             self.label_map[char] = i

     def encode(self, text):
         encoded_label = []
         for i, char in enumerate(text):
             if i > 0 and char == text[i - 1]:
                 encoded_label.append(0)    # 같은 문자 사이에 공백 문자 label을 삽입
             encoded_label.append(self.label_map[char])
         return np.array(encoded_label, dtype=np.int32)

     def decode(self, encoded_label):
         target_characters = list(self.character)
         decoded_label = ""
         for encode in encoded_label:
             decoded_label += self.character[encode]
         return decoded_label

def DatasetLabelEncoder(dataset, target):
    result = dict()
    for i in range(len(dataset)):
        _, t = dataset[i]
        for d in t:
            if d[target] not in result:
                result[d[target]] = len(result)
    return result

def ModelFromhdf5(path):
    with h5py.File(path, "r") as f:
        print("=== root keys ===")
        print(list(f.keys()))
        print()

        # 1) layer_names 읽기 (bytes / str 둘 다 대응)
        if "layer_names" in f.attrs:
            raw_layer_names = f.attrs["layer_names"]
            layer_names = [
                n.decode("utf-8") if isinstance(n, (bytes, bytearray)) else str(n)
                for n in raw_layer_names
            ]
            print("레이어 이름들:", layer_names)

            for layer_name in layer_names:
                g = f[layer_name]
                print(f"\n[레이어] {layer_name}")

                # 2) weight_names 도 마찬가지로 bytes/str 둘 다 처리
                if "weight_names" in g.attrs:
                    raw_weight_names = g.attrs["weight_names"]
                    weight_names = [
                        wn.decode("utf-8") if isinstance(wn, (bytes, bytearray)) else str(wn)
                        for wn in raw_weight_names
                    ]
                    for wn in weight_names:
                        dset = g[wn]
                        print(f"  {wn} : shape={dset.shape}")
                else:
                    for sub in g.keys():
                        dset = g[sub]
                        print(f"  {layer_name}/{sub} : shape={dset.shape}")

        else:
            print("attrs['layer_names'] 가 없습니다.")
            for k in f.keys():
                print("group:", k)
                if isinstance(f[k], h5py.Group):
                    print("  subkeys:", list(f[k].keys()))
    f.close()

def _find_bilstm_cell_groups(h5_group):
    cell_paths = {}
    def visitor(name, obj):
        if not isinstance(obj, h5py.Dataset):
            return
        if not name.endswith("kernel:0"):
            return

        lower = name.lower()
        # name 예: "forward_lstm/lstm_cell_1/kernel:0"
        group_path = name.rsplit("/", 1)[0]

        if "forward" in lower and "forward" not in cell_paths:
            cell_paths["forward"] = group_path
        elif ("backward" in lower or "reverse" in lower) and "backward" not in cell_paths:
            cell_paths["backward"] = group_path

    h5_group.visititems(visitor)

    result = {}
    for direction, path in cell_paths.items():
        result[direction] = h5_group[path]
        
    return result

def load_one_direction(cell_group, module, hidden_size, suffix):
    # Keras: kernel: (in_dim, 4*h), recurrent_kernel: (h, 4*h)
    kernel = torch.from_numpy(cell_group["kernel:0"][()]).to(device)
    recurrent = torch.from_numpy(cell_group["recurrent_kernel:0"][()]).to(device)
    
    W_ih = kernel.t()      # (4*h, in_dim)
    W_hh = recurrent.t()   # (4*h, h)
    
    w_ih_attr = "weight_ih_l0" + suffix
    w_hh_attr = "weight_hh_l0" + suffix
    b_ih_attr = "bias_ih_l0" + suffix
    b_hh_attr = "bias_hh_l0" + suffix
    
    w_ih_param = getattr(module, w_ih_attr)
    w_hh_param = getattr(module, w_hh_attr)
    
    with torch.no_grad():
        w_ih_param.copy_(W_ih)
        w_hh_param.copy_(W_hh)
    
    # bias
    if "bias:0" in cell_group and getattr(module, b_ih_attr) is not None:
        b = torch.from_numpy(cell_group["bias:0"][()]).to(device)  # (4*h,) or (8*h,)
    
        if b.numel() == 4 * hidden_size:
            b_ih = b
            b_hh = torch.zeros_like(b)
        elif b.numel() == 8 * hidden_size:
            b_ih = b[:4 * hidden_size]
            b_hh = b[4 * hidden_size:]
    
        b_ih_param = getattr(module, b_ih_attr)
        b_hh_param = getattr(module, b_hh_attr)
    
        with torch.no_grad():
            b_ih_param.copy_(b_ih)
            b_hh_param.copy_(b_hh)

def LoadModelFromhdf5(model, path, mapping, device):
    with h5py.File(path, "r") as f:            
        for name, module in model.named_children():
            if name in list(mapping.keys()):
                group_name = mapping[name]
                h5_group = f[group_name]
                if isinstance(module, (nn.Linear, nn.Conv2d)):
                    param_name_candidates = {
                        "weight": ["kernel:0", "weight"],
                        "bias":   ["bias:0", "bias"],
                    }
                    for p_name, param in module.named_parameters(recurse=False):
                        candidates = param_name_candidates.get(p_name, [p_name])
    
                        for ds_name in candidates:
                            w = torch.from_numpy(h5_group[group_name+"/"+ds_name][()]).to(device)
                            if isinstance(module, nn.Linear):
                                # Keras Dense: (in_dim, out_dim)
                                # PyTorch Linear: (out_dim, in_dim)
                                if w.ndim == 2 and w.shape[::-1] == param.shape:
                                    w = w.t()
                                    
                            elif isinstance(module, nn.Conv2d):
                                # Keras Conv2D: (kh, kw, in_ch, out_ch)
                                # PyTorch Conv2d: (out_ch, in_ch, kh, kw)
                                if w.ndim == 4 and w.shape[2] == param.shape[1]:
                                    w = w.permute(3, 2, 0, 1)
                                    
                            with torch.no_grad():
                                param.copy_(w)

                            break
                            
                elif isinstance(module, nn.LSTM):
                    hidden_size = module.hidden_size
                    cell_groups = _find_bilstm_cell_groups(h5_group)
                    fwd_group = cell_groups.get("forward")
                    bwd_group = cell_groups.get("backward")

                    load_one_direction(fwd_group, module, hidden_size, suffix="")
                    load_one_direction(bwd_group, module, hidden_size, suffix="_reverse")
                else:
                    continue
            else:
                continue
    return model

def editDistRec(s1, s2, m, n):
    if m == 0:
        return n

    if n == 0:
        return m

    if s1[m - 1] == s2[n - 1]:
        return editDistRec(s1, s2, m - 1, n - 1)

    prev = list(range(n + 1))
    curr = [0] * (n + 1)
    
    for i in range(1, m + 1):
        curr[0] = i
        ai = s1[i - 1]
        for j in range(1, n + 1):
            cost = 0 if ai == s2[j - 1] else 1
            curr[j] = min(
                prev[j] + 1,        # 삭제
                curr[j - 1] + 1,    # 삽입
                prev[j - 1] + cost, # 치환
            )
        prev, curr = curr, prev

    return prev[n]

def editDistance(s1, logits, label_length, label_converter):
    pred_texts = check_inference_with_logit(logits, label_converter=label_converter, full_texts=True)
    labels_padded = s1.detach().cpu()
    label_length = label_length.detach().cpu().tolist()
    gt_texts = []
    for b, L in enumerate(label_length):
        L = int(L)
        # padding 제거 후 decode
        seq = labels_padded[b, :L].tolist()
        gt_str = label_converter.decode(seq).replace('-', '')
        gt_texts.append(gt_str)

    total_ed = 0
    total_len = 0
    for pred, gt in zip(pred_texts, gt_texts):
        ed = editDistRec(pred, gt, len(pred), len(gt))
        total_ed += ed
        total_len += len(gt)
            
    avg_ed  = total_ed / max(total_len, 1)
    return avg_ed
```

```python
# Dataset 설계
#=======================================================================================================================================================================
class MJDataset(Dataset):
    # 객체를 초기화 할 때 lmdb를 열어 env에 준비해둡니다
    # 또, lmdb에 있는 데이터 수를 미리 파악해둡니다
    def __init__(
        self,
        dataset_path,
        label_converter,
        img_size=(100, 32),
        max_text_len=22,
        character="",
        max_sample=None,
        transform = None
    ):
        super().__init__()
        self.label_converter = label_converter
        self.img_size = img_size
        self.max_text_len = max_text_len
        self.character = character
        self.dataset_path = dataset_path
        self.labels = []
        self.encoded_labels = []
        
        # lmdb open
        self.env = lmdb.open(dataset_path,max_readers=32,readonly=True,lock=False,readahead=False,meminit=False)
        with self.env.begin(write=False) as txn:
            self.num_samples = int(txn.get("num-samples".encode()))
            self.index_list = [idx + 1 for idx in range(self.num_samples)]

        if max_sample is not None:
            data_range = min(self.num_samples, max_sample)
        
        for i in tqdm(range(data_range)):
            index = self.index_list[i]
            with self.env.begin(write=False) as txn:
                label_key = f"label-{index:09d}".encode()
                label = txn.get(label_key).decode("utf-8")
    
                img_key = f"image-{index:09d}".encode()
                imgbuf = txn.get(img_key)
    
                buf = six.BytesIO()
                buf.write(imgbuf)
                buf.seek(0)
    
                try:
                    img_pil = Image.open(buf).convert("RGB")
                except IOError:
                    label = "-"
    
            # 레이블 전처리
            label = label.upper()
            out_of_char = f"[^{self.character}]"
            label = re.sub(out_of_char, "", label)
            label = label[: self.max_text_len]
            encoded_label = self.label_converter.encode(label)
            
            self.labels.append(label)
            self.encoded_labels.append(encoded_label)

        self.max_len = len(max(self.encoded_labels, key=len))
        
    def __len__(self):
        return len(self.labels)

    # index에 해당하는 image와 label을 읽어옵니다
    # 위에서 사용한 코드와 매우 유사합니다
    # label을 조금 더 다듬는 것이 약간 다릅니다
    def __getitem__(self, idx):
        index = self.index_list[idx]
        with self.env.begin(write=False) as txn:
            label_key = f"label-{index:09d}".encode()
            label = txn.get(label_key).decode("utf-8")

            img_key = f"image-{index:09d}".encode()
            imgbuf = txn.get(img_key)

            buf = six.BytesIO()
            buf.write(imgbuf)
            buf.seek(0)

            try:
                img_pil = Image.open(buf).convert("RGB")
            except IOError:
                img_pil = Image.new("RGB", self.img_size)
                label = "-"

        orig_w, orig_h = img_pil.size
        target_width = min(int(orig_w * self.img_size[1] / orig_h), self.img_size[0])
        target_img_size = (target_width, self.img_size[1])
        img_pil = img_pil.resize(target_img_size)

        img = np.array(img_pil)
        img = img.transpose(2, 0, 1)

        padded_img = np.zeros((3, self.img_size[1], self.img_size[0]), dtype=np.float32)
        c, h, w = img.shape
        padded_img[:, :h, :w] = img
        
        label = self.labels[idx]
        encoded_label = self.encoded_labels[idx]
        encoded_label = np.asarray(encoded_label, dtype=np.int64)
        label_len = len(encoded_label)
        label_padded = np.full((self.max_len,), 0, dtype=np.int64)
        truncated_len = min(label_len, self.max_len)
        label_padded[:truncated_len] = encoded_label[:truncated_len]
        input_length = np.int64(24)
        
        return padded_img, label_padded, input_length, truncated_len

class CAPTCHADataset(Dataset):
    # 객체를 초기화 할 때 lmdb를 열어 env에 준비해둡니다
    # 또, lmdb에 있는 데이터 수를 미리 파악해둡니다
    def __init__(
        self,
        dataset_path,
        label_converter,
        img_size=(100, 32),
        max_text_len=22,
        character="",
    ):
        super().__init__()
        self.label_converter = label_converter
        self.img_size = img_size
        self.max_text_len = max_text_len
        self.character = character
        self.paths = []
        self.labels = []
        extensions = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tif", ".tiff"}
        for dirpath, dirnames, filenames in os.walk(dataset_path):
            for fname in filenames:
                ext = os.path.splitext(fname)[1].lower()
                if ext in extensions:
                    path = os.path.join(dirpath, fname)
                    label = os.path.splitext(fname)[0]  # 확장자 제거
                    out_of_char = f"[^{self.character}]"
                    label = re.sub(out_of_char, "", label)
                    label = label[: self.max_text_len].upper()
                    self.paths.append(path)
                    self.labels.append(label)

    def __len__(self):
        return len(self.paths)

    # index에 해당하는 image와 label을 읽어옵니다
    # 위에서 사용한 코드와 매우 유사합니다
    # label을 조금 더 다듬는 것이 약간 다릅니다
    def __getitem__(self, idx):
        path = self.paths[idx]
        try:
            img_pil = Image.open(path).convert("RGB")
        except IOError:
            img_pil = Image.new("RGB", self.img_size)
            label = "-"

        orig_w, orig_h = img_pil.size
        target_width = min(int(orig_w * self.img_size[1] / orig_h), self.img_size[0])
        target_img_size = (target_width, self.img_size[1])
        img_pil = img_pil.resize(target_img_size)

        img = np.array(img_pil)
        img = img.transpose(2, 0, 1)

        padded_img = np.zeros((3, self.img_size[1], self.img_size[0]), dtype=np.float32)
        c, h, w = img.shape
        padded_img[:, :h, :w] = img

        label = self.labels[idx]
        encoded_label = self.label_converter.encode(label)

        return padded_img, encoded_label, len(encoded_label), label

class WaterMeterDataset(Dataset):
    # 객체를 초기화 할 때 lmdb를 열어 env에 준비해둡니다
    # 또, lmdb에 있는 데이터 수를 미리 파악해둡니다
    def __init__(
        self,
        dataset_path,
        label_converter,
        img_size=(100, 32),
        max_text_len=22,
        character="",
    ):
        super().__init__()
        self.label_converter = label_converter
        self.img_size = img_size
        self.max_text_len = max_text_len
        self.character = character
        self.data = pd.read_csv(os.path.join(dataset_path,'data.csv'))
        self.dataset_path = dataset_path

    def __len__(self):
        return len(self.data)

    # index에 해당하는 image와 label을 읽어옵니다
    # 위에서 사용한 코드와 매우 유사합니다
    # label을 조금 더 다듬는 것이 약간 다릅니다
    def __getitem__(self, idx):
        img_name = self.data['photo_name'][idx]
        path = os.path.join(self.dataset_path, 'images',img_name)
        try:
            img_pil = Image.open(path).convert("RGB")
        except IOError:
            img_pil = Image.new("RGB", self.img_size)
            label = "-"

        orig_w, orig_h = img_pil.size
        target_width = min(int(orig_w * self.img_size[1] / orig_h), self.img_size[0])
        target_img_size = (target_width, self.img_size[1])
        img_pil = img_pil.resize(target_img_size)

        img = np.array(img_pil)
        img = img.transpose(2, 0, 1)

        padded_img = np.zeros((3, self.img_size[1], self.img_size[0]), dtype=np.float32)
        c, h, w = img.shape
        padded_img[:, :h, :w] = img

        label = str(self.data['value'][idx])
        encoded_label = self.label_converter.encode(label)

        return padded_img, encoded_label, len(encoded_label), label

def collate_fn(batch):
    imgs, encoded_labels, label_lens, raw_labels = zip(*batch)

    imgs_tensor = torch.tensor(np.stack(imgs, axis=0), dtype=torch.float32)

    max_len = max(label_lens)
    labels_padded = torch.zeros(len(batch), max_len, dtype=torch.long)
    for i, label_arr in enumerate(encoded_labels):
        length = label_lens[i]
        labels_padded[i, :length] = torch.tensor(label_arr, dtype=torch.long)

    batch_size = imgs_tensor.size(0)
    # 여기서는 크기 24로 가정
    input_length = torch.full(size=(batch_size,), fill_value=24, dtype=torch.long)
    label_length = torch.tensor(label_lens, dtype=torch.long)

    return (
        imgs_tensor,
        labels_padded,
        input_length,
        label_length,
    )
```

```python
# 학습 함수
#=======================================================================================================================================================================
def train_model(model, train_dataset, label_converter, train_ratio=0.8, valid_dataset=None,
                criterion = 'CE', optimize='SGD', batch_size=256, collate_fn = None,
                epochs=50, lr = 0.001, verbose = 1, isTrained = False, 
                random_state=None, view_accurate_value = False,
                epoch_shuffle=False, shuffle_epoch=1, shuffle_len=131072, num_workers=None, 
                transform=None, transform_p=1.0, scheduler=None, warmup_steps=3000,
                drop_last=False, patience=4, model_name="Model1"):

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    if isTrained:
        model.to(device)
    else:
        model = model(
            dropout=DROPOUT
        )
    
        model.to(device)
        
    # 손실 함수
    if criterion=="CE":
        criterion = nn.CrossEntropyLoss()
    elif criterion=="BCE":
        criterion = nn.BCEWithLogitsLoss()
    elif criterion=="CTC":
        criterion = nn.CTCLoss(blank=0, reduction='mean', zero_infinity=True)

    # 옵티마이저
    if optimize=='AdamW':
        optimizer=optim.AdamW(model.parameters(), lr=lr)
    elif optimize=='Adam':
        optimizer=optim.Adam(model.parameters(), lr=lr)
    elif optimizer=="SGD":
        optimizer=optim.SGD(model.parameters(), lr=lr)
    elif optimizer=="Adadelta":
        optimizer = optim.Adadelta(model.parameters(), lr=lr)
    else:
        optimizer=optim.Adam(model.parameters(), lr=lr)
    
    train_size = int(len(train_dataset) * train_ratio)
    val_size = len(train_dataset) - train_size
    if valid_dataset is None:
        if random_state is None:
            # 랜덤하게 dataset나누기
            tr_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])
        else:
            manual_seed = random_state
            generator = torch.Generator().manual_seed(manual_seed)
        
        # 랜덤하게 dataset나누기
        tr_dataset, val_dataset = random_split(train_dataset, [train_size, val_size], generator=generator)
        if num_workers is not None:
            val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, drop_last=True, collate_fn=collate_fn, num_workers=num_workers, pin_memory=True, persistent_workers=True)
        else:
            val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, drop_last=True, collate_fn=collate_fn)
    else:
        tr_dataset = train_dataset
        train_size = len(train_dataset)
        if num_workers is not None:
            val_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, drop_last=True, collate_fn=collate_fn, num_workers=num_workers, pin_memory=True, persistent_workers=True)
        else:
            val_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, drop_last=True, collate_fn=collate_fn)

    if (scheduler is not None) and scheduler=="Warmup_cosine":
        sched_warmup = LinearLR(optimizer, start_factor=1e-3, end_factor=1.0, total_iters=max(1, warmup_steps))
        sched_cosine = CosineAnnealingLR(optimizer, T_max=max(1, train_size - warmup_steps), eta_min=1e-6)
        scheduler_ = SequentialLR(optimizer, schedulers=[sched_warmup, sched_cosine], milestones=[warmup_steps])
    elif scheduler=="OneCycleLR":
        scheduler_ = OneCycleLR(optimizer, max_lr=3e-4, steps_per_epoch=train_size, epochs=epochs, pct_start=0.1, anneal_strategy='cos', div_factor=25.0, final_div_factor=1e3)
    elif scheduler=="MultiStepWarmUpLR":
        steps_per_epoch = train_size//batch_size
        scheduler_ = MultiStepWarmUpLR(initial_learning_rate=1e-2, lr_steps=[e * steps_per_epoch for e in [50, 70]], lr_rate=0.1, warmup_steps=5 * steps_per_epoch, min_lr=1e-4)
    else:
        scheduler_ = None

    train_losses = []
    val_losses = []
    train_edit_distance_list = []
    val_edit_distance_list = []
    
    #train_loader = apply_cutmix_mixup_on_dataset(tr_dataset, batch_size=batch_size, is_test=is_test, with_cutmix=with_cutmix, with_mixup=with_mixup,num_classes=num_classes)
    model_hist = dict()
    best_val_loss = float('inf')
    best_param = dict()

    if train_size%batch_size == 0:
        add_batch = 0
    else:
        add_batch = 1
    
    
    model.train() 
    for epoch in range(epochs):
        total_loss, total_edit_distance= 0, 0
        # PyTorch DataLoader 설정. epoch 마다 섞이게 만듦, 확률에 따라 trainset에 augment 적용
        if num_workers is not None:
            if epoch_shuffle:
                if epoch % shuffle_epoch==0:
                    tr_dataset, _ = random_split(train_dataset, [shuffle_len, len(train_dataset)-shuffle_len])
                    train_size = len(tr_dataset)
                    if train_size%batch_size == 0:
                        add_batch = 0
                    else:
                        add_batch = 1
                train_loader = DataLoader(tr_dataset, batch_size=batch_size, shuffle=True, drop_last=True, collate_fn=collate_fn, num_workers=num_workers, pin_memory=True, persistent_workers=True)
            else:
                train_loader = DataLoader(tr_dataset, batch_size=batch_size, shuffle=False, drop_last=True, collate_fn=collate_fn, num_workers=num_workers, pin_memory=True, persistent_workers=True)
        else:
            if epoch_shuffle:
                if epoch % shuffle_epoch==0:
                    tr_dataset, _ = random_split(train_dataset, [shuffle_len, len(train_dataset)-shuffle_len])
                    train_size = len(tr_dataset)
                    if train_size%batch_size == 0:
                        add_batch = 0
                    else:
                        add_batch = 1
                train_loader = DataLoader(tr_dataset, batch_size=batch_size, shuffle=False, drop_last=True, collate_fn=collate_fn)
            else:
                train_loader = DataLoader(tr_dataset, batch_size=batch_size, shuffle=True, drop_last=True, collate_fn=collate_fn)

        for step, batch in zip(tqdm(range(train_size//batch_size+add_batch)), train_loader):
            optimizer.zero_grad()
            loss, edit_distance = train_step(model, batch, optimizer, criterion, label_converter, device)
            total_loss += loss
            total_edit_distance += edit_distance

        avg_loss = total_loss / len(train_loader)
        avg_edit_distance = total_edit_distance / len(train_loader)
        train_losses.append(avg_loss)
        train_edit_distance_list.append(avg_edit_distance)

        # Validation loss 계산
        model.eval()
        val_loss, val_edit_distance = 0, 0
        with torch.no_grad():
            for batch in val_loader:
                loss, edit_distance = eval_step(model, batch, criterion, label_converter, device)

                val_loss += loss
                val_edit_distance += edit_distance

        val_loss /= len(val_loader)
        val_edit_distance /=len(val_loader)
        val_losses.append(val_loss)
        val_edit_distance_list.append(val_edit_distance)

        # Early Stopping 조건
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_param = copy.deepcopy(model.state_dict())
            early_stop_counter = 0
        else:
            early_stop_counter += 1

        if early_stop_counter >= patience:
            best_param = copy.deepcopy(model.state_dict())
            early_stop_counter = -float('inf')

       
        if (epoch+1)%verbose==0:
            if view_accurate_value:
                print(f"{model_name} | Epoch {epoch+1}/{epochs} | Train Loss: {avg_loss}, Train Edit Distance: {avg_edit_distance}| Val Loss: {val_loss}, Val Edit Distance: {val_edit_distance}")
            else:
                print(f"{model_name} | Epoch {epoch+1}/{epochs} | Train Loss: {avg_loss:.4f}, Train Edit Distance: {avg_edit_distance:.4f} | Val Loss: {val_loss:.4f}, Val Edit Distance: {val_edit_distance:.4f}")
        gc.collect()
        torch.cuda.empty_cache()
        model.train()
        
    model.eval()
    model.load_state_dict(best_param)
   
    if not isTrained:
        return model, train_losses, val_losses
    else:
        return train_losses, val_losses, train_edit_distance_list, val_edit_distance_list

def train_step(model, batch, optimizer, loss_function, label_converter, device):
    model.train()
    imgs, labels_padded, input_lengths, label_lengths = [x.to(device) for x in batch[:4]]

    optimizer.zero_grad()

    # 모델 포워드 패스
    logits  = model(imgs) # (T,B,C)

    loss = loss_function(logits, labels_padded, input_lengths, label_lengths)

    # Backpropagation
    loss.backward()
    optimizer.step()
    
    return loss.item() , editDistance(labels_padded, logits, label_lengths, label_converter=label_converter)
    
def eval_step(model, batch, loss_function, label_converter, device):
    model.eval()
    
    imgs, labels_padded, input_lengths, label_lengths = [x.to(device) for x in batch[:4]]

    # 모델 포워드 패스
    logits  = model(imgs) # (T,B,C)

    loss = loss_function(logits, labels_padded, input_lengths, label_lengths)

    return loss.item(),editDistance(labels_padded, logits, label_lengths, label_converter=label_converter)
```

```python
#inference
#================================================================================
def decode_greedy(output, label_converter):
    # (T,B,C) -> (B,T) index
    out = output.detach().cpu().numpy()  # (T,B,C)
    argmax = out.argmax(axis=2).transpose()  # (B,T)

    results = []
    for seq in argmax:
        # 연속된 동일 글자(또는 blank=0) 제거 로직을 적용해야
        # CTC 디코딩다운 결과가 나옵니다.
        # 여기서는 간단히 blank(0) 무시하고 연속 제거만 보여줌
        decoded = []
        prev = None
        for idx in seq:
            if idx != 0 and idx != prev:
                decoded.append(idx)
            prev = idx
        # 인덱스를 실제 문자로
        decoded_str = label_converter.decode(decoded).replace('-', '')
        results.append(decoded_str)
    return results

def check_inference(model, dataset, index=5):
    for i in range(index):
        img, encoded_label, label_len, raw_label = dataset[i]  # 단일 데이터
        img_tensor = torch.tensor(img[np.newaxis, ...], dtype=torch.float32).to(device)  # (1,3,32,100)
        output = model(img_tensor)  # (T,1,num_chars)
        # 디코딩
        result_text = decode_greedy(output, label_converter)[0]
        print(f"GT: {raw_label} / Pred: {result_text}")

        # 이미지를 시각화하고 싶으면
        display(Image.fromarray(img.transpose(1,2,0).astype(np.uint8)))

def check_inference_with_logit(logit, label_converter = None, index=5, full_texts=False):
    if label_converter is not None:
        if full_texts:
            result_text = decode_greedy(logit, label_converter)
        else:
            result_text = decode_greedy(logit, label_converter)
        return result_text
    else:
        return ""
        
def check_inference_with_pretrained(dataset, index=5):
    for i in range(index):
        img_pil, cropped_imgs, ocr_result = detect_text_with_dataset(dataset, i)
        for idx, word_img in enumerate(cropped_imgs):
            print(f"[{idx}] ", end="")
            recognize_img(word_img)
```

```python
# 태스트 함수
#=======================================================================================================================================================================
def detect_text(img_path):
    # PIL로 이미지 열기
    img_pil = Image.open(img_path).convert('RGB')
    # EasyOCR은 numpy 배열을 입력받으므로 변환
    img_np = np.array(img_pil)

    # EasyOCR을 이용한 텍스트 검출 및 인식
    # 결과 형식: [(bbox, text, confidence), ...]
    ocr_result = reader.readtext(img_np)

    # 시각화를 위해 ImageDraw 객체 생성
    img_draw = ImageDraw.Draw(img_pil)
    cropped_imgs = []

    for bbox, text, conf in ocr_result:
        # 좌표를 정수형으로 변환
        bbox_int = [(int(x), int(y)) for x, y in bbox]
        img_draw.polygon(bbox_int, outline='red')

        xs = [pt[0] for pt in bbox_int]
        ys = [pt[1] for pt in bbox_int]
        x_min = max(0, min(xs) - 5)
        y_min = max(0, min(ys) - 5)
        x_max = max(xs) + 5
        y_max = max(ys) + 5
        word_box = (x_min, y_min, x_max, y_max)
        cropped_imgs.append(img_pil.crop(word_box))

    return img_pil, cropped_imgs, ocr_result

def detect_text_with_dataset(dataset, idx):
    # PIL로 이미지 열기
    img_pil, _, _, _ = dataset[idx] 
    img_pil = np.transpose(img_pil, (1, 2, 0))
    # EasyOCR은 numpy 배열을 입력받으므로 변환
    img_np = np.array(img_pil)
    img_pil = Image.fromarray(img_np.astype(np.uint8))

    # EasyOCR을 이용한 텍스트 검출 및 인식
    # 결과 형식: [(bbox, text, confidence), ...]
    ocr_result = reader.readtext(img_np)

    # 시각화를 위해 ImageDraw 객체 생성
    img_draw = ImageDraw.Draw(img_pil)
    cropped_imgs = []

    for bbox, text, conf in ocr_result:
        # 좌표를 정수형으로 변환
        bbox_int = [(int(x), int(y)) for x, y in bbox]
        img_draw.polygon(bbox_int, outline='red')

        xs = [pt[0] for pt in bbox_int]
        ys = [pt[1] for pt in bbox_int]
        x_min = max(0, min(xs) - 5)
        y_min = max(0, min(ys) - 5)
        x_max = max(xs) + 5
        y_max = max(ys) + 5
        word_box = (x_min, y_min, x_max, y_max)
        cropped_imgs.append(img_pil.crop(word_box))

    display(img_pil)

    return img_pil, cropped_imgs, ocr_result

def recognize_img(pil_img, input_img_size=(100, 32)):
    # TODO: 잘려진 단어 이미지를 인식하는 코드를 작성하세요!

    # PIL 이미지를 numpy 배열로 변환
    img_np = np.array(pil_img)
    # 인식 모델 입력 크기에 맞게 이미지 크기 조정
    img_resized = cv2.resize(img_np, input_img_size)

    # detail=0 옵션은 인식 결과의 텍스트만 반환합니다.
    result = reader.readtext(img_resized, detail=0, paragraph=False)

    if result:
        recognized_text = result[0]
    else:
        recognized_text = ""

    print("Recognized text:", recognized_text)
    return recognized_text
```

OCR은 Optical Character Recognition으로 다음과 같이 여러 문자를 인삭하는 것을 의미합니다. 



이러한 OCR은 흔히 번호판 인식, 비대면 신분 증명 등에서 활용되곤 합니다. 특히 시각적으로 불편하신 분들을 위한 OCR 기술들도 개발되고 있습니다. 따라서, OCR 기술에 대한 이해하기 위해 이론적인 측면이 아닌 실험적인 측면과 더욱 나아가 어떤 방향성으로 나아가야 하는지에 대한 문제 파악을 해야 한다고 생각하였습니다. 위의 실험 결과는 실험적인 측면으로 baseline 부터 가장 초기의 OCR 모델인 RCNN모델을 학습시켜 보고, 모델의 성능을 확인하였으며, pretrained 모델과 실제 배포된 모델들로 실험해 보았습니다. 또한, 여러 dataset을 구성하여 해당 dataset에 대해서도 잘 대응 할 수 있는지를 확인해 보았습니다. 준비한 dataset은 watermeter dataset과 capcha dataset으로 우리가 일상생활에서 만날 수 있는 문제를 vision recognition으로 해결하기 위하여 준비된 dataset입니다.


# 결과




RCNN모델을 MJ dataset으로 학습시킨 결과 loss가 빠르게 줄어들었습니다. 학습에 활용된 data는 2^20 = 1,048,576 개 였으며 validation set에는 불필요하게 과도한 검증을 하지 않기 위하여 train set과의 비율을 조정 해 2^16 = 65536 개를 사용하였고 총 2시간정도 소요되었습니다. 또한, 여러 병목 현상을 발견하여 이를 해결하기 위해 기본 코드에 변경을 가했습니다. 





loss의 측면에서 보면 첫 epoch 때 3.0788 에서 시작한 모델이 점점 loss가 감소하여 4 epoch 때 0.5192 까지 감소하였습니다. 또한, 그래프를 살펴보면 4 epoch 까지가 가장 급격한 변화이고 이 이후에는 validation set과 train set 모두에서 특정 값에서 크게 변동하지 않는 모습을 보였기 때문에 수렴했다고 판단하였지만 이후 data의 변화를 보기 위하여 추가적으로 20 epoch 까지 수행하였습니다. overfitting의 경향성은 9 epoch에서 나타났으며 Val Loss: 0.2187 와 Val Edit Distance: 0.3274 로 가장 낮은 수치를 기록하였습니다.





Edit distance는 삽입/삭제/치환과 같은 편집으로 몇번의 편집으로 두 문자가 같아질 수 있는지를 측정하는 matric으로 두 문자의 유사도를 판단합니다. 즉, logit으로 부터 decode_greedy로 문자열의 최종값을 얻어 label과 같이 padding을 재거하고 두 문자의 유사도를 비교한 후 평균을 구하였습니다. 즉, 현재 문자열에서 평균적으로 몇 번의 편집이 필요한지를 정량적으로 나타냅니다. 이번 실험에서 edit distance 는 validation set 기준 0.37이 나왔으며 이는 100개의 문자를 본다고 가정하였을 때 평균적으로 37번의 편집을 거쳐야 한다는 의미가 됩니다.


따라서, 결과를 바탕으로 해석하였을 때 22개의 문자를 가장 긴 문자로 설정하였음으로 validation loss와 edit distance가 동시에 낮았던 9 epoch 기준으로 계산한다면 7.04 번의 편집이 필요하다는 의미가 됩니다. 

해당 모델을 기반으로 test set의 data를 실험하였을 때는 상당히 만족스러운 결과를 얻을 수 있었습니다.





하지만, 다른 dataset인 capcha dataset은 만족스럽지 못한 결과를 보여주었습니다. 이는, 학습때와는 다르게 model이 인식하지 못 할만한 요소들을 넣은 고난이도 data인 capcha data를 사용한 것이 이유일 수 있습니다. 하지만, 모델이 학습을 시켜도 비슷한 dataset을 잘 구분하지 못하는 지, 즉 모델의 한계인지, 아니면 학습의 부족인 지 등은 자세한 실험이 추가적으로 필요합니다.




또 다른 dataset인 watermeter dataset의 경우에도 실망스러운 결과를 반환하였습니다. 여기서 주목한 점은 모델이 숫자가 아닌 글자를 주로 인식하였다는 점입니다. 데이터 자체의 resolution이 낮아진 점 도 있겠지만, 이전의 capcha dataset에서도 비슷한 문제가 있었습니다. 따라서, 모델이 학습한 데이터가 숫자와 문자가 섞인 data에 대하여 편향되게 학습했을 가능성이 있습니다. 이 부분에 대해서는 추가적인 실험이 필요합니다.




미리 배포된 모델인 EasyOCR 대해서도 확인할 수 있었습니다. 이는 학습된 모델들과 달리 높은 성능을 보여주었습니다. 따라서, 이전에 인식하지 못한 dataset의 data를 인식시킨 결과는 다른 insight를 주었습니다. 





기본적인 test dataset에 대해서는 높은 정확도를 보여주었습니다. 하지만, capcha dataset은 실망스러운 결과를 반환한 것은 마찬가지였습니다. 하지만, 특정 case에 대하여 맞추는 경우가 발생하여 성능을 추가적으로 실험해야 합니다. watermeter dataset에 대하여는 감지하지 못하는 경향성을 보여 해당 dataset에 대해서는 낮은 resolution에 대한 문제로 의심해 볼 수 있을 것 같습니다.

dataset에 대하여 dataloader에서 큰 병목이 발생하여 다음과 같이 dataset의 길이를 조정해 가며 batch_size=2048를 dataloader에서 꺼내오는 대 얼마 만큼의 시간이 걸리는지 확인하였습니다. 결과는 아래와 같습니다.

|number of data|time(1 batch)(s)|time(total)(s)|
|-----|--------|--------|
|8192|4.5531|32.9303|
|16384|4.7979|71.5162|
|32768|5.4007|144.1839|
|65536|4.9647|297.0061|

모든 dataset의 길이는 2의 재곱으로 표현되며 길이를 2배 늘렸을 때 시간도 거의 2배 가량 증가하였습니다.

위의 결과를 바탕으로 가용할 수 있는 시간을 기준으로 생각하였을 때 65536 개의 dataloader이 불러와 지는 대 약 5분이고 131072 개가 불러와 지는 대 10분의 시간이 걸릴 것으로 예측됩니다. 실제 측정 결과가 기록되지는 않았지만 실제 dataloader에서 data가 불러와 지는데 약 10 정도가 소요되었습니다.

따라서, dataset을 메모리의 허용량 까지 올리는 방식을 채택하였습니다. 이는 lmdb와 dataloader의 num_worker 옵션을 같이 사용하였을 때, lmdb가 파일을 열어 둔 상태라면 "TypeError: can't pickle Environment objects" 오류가 나타나기 때문에 lmdb이 environment object를 사용하고 있어서 문제가 발생하는 것이라는 추측에서 시작하였습니다. 해당 추측에 기반하여 __getitem__으로 가져오는 item을 내부 list에서 가져오도록 설정하였습니다.

하지만 결과는 차이가 없었습니다. 같은 dataset의 길이인 65536에 대하여 비슷한 시간이 소요되었습니다. 따라서, dataset에서의 문제보다 collate_fn에서의 문제라고 판단하였습니다.

collate_fn의 기능을 batch 단위로 가장 긴 encoding된 label의 길이에 맞추는 기능을 수행합니다. 즉, encoding된 label의 길이도 미리 맞추어주면 collate_fn이 없이도 dataloader가 구성이 가능합니다. 기능적으로는 collate_fn으로 수행하는 것이 맞으나 시간 자원 측면에서 collate_fn을 포기하는 것이 더 좋다고 판단하였습니다.

collate_fn을 포기한 결과 2^20=1048576개의 data에 대하여 학습하는 데 1 epoch 가 약 6분 50초 정도로 단축되었습니다. batch size만큼의 data를 꺼내와 연산하는 것은 batch size만큼의 data를 꺼내오는 과정을 포함함으로 batch에서 data를 꺼내오는 것은 6분 50초 이하일 것으로 예상됩니다. 만약, 기존의 방식을 유지했다면 dataset의 길이가 2배가 늘어났을 때 시간이 약 2배 늘어났음으로, 2^20=1048576개의 data를 꺼내오는 대 2^16 = 65536 에서 16배의 시간이 늘어난 1시간 19분 정도가 예상됩니다.
